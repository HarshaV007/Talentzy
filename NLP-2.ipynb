```python
#Exxtracting Minimum Experience required by the company for the posted job
df['Min_Experience']=df['Experience']
df['Min_Experience']=df['Min_Experience'].str.split('-').str[0]
df['Min_Experience']=df['Min_Experience'].str.split(' ').str[1]
df['Min_Experience']=df['Min_Experience'].fillna('0')
df['Min_Experience']=df['Min_Experience'].astype('int')


#Maximum Experience can also be extracted if very specific demand is to be fullfilled
#df['Max_Experience']=df['Experience'].str.split()[1]
```

```python
df.dtypes
```

```python
df['Technical_matchings']=df['Technical_matchings'].str.split("'")
df['Generic_matchings']=df['Generic_matchings'].str.split("'")

```

```python
#Extracting skills

df['Technical_Skill1']=df['Technical_matchings'].str[1]
df['Technical_Skill2']=df['Technical_matchings'].str[3]
df['Technical_Skill3']=df['Technical_matchings'].str[5]
df['Technical_Skill4']=df['Technical_matchings'].str[7]
df['Technical_Skill5']=df['Technical_matchings'].str[9]
df['Technical_Skill6']=df['Technical_matchings'].str[11]
df['Technical_Skill7']=df['Technical_matchings'].str[13]
df['Technical_Skill8']=df['Technical_matchings'].str[15]
df['Technical_Skill9']=df['Technical_matchings'].str[17]
df['Technical_Skill10']=df['Technical_matchings'].str[19]
df['Generic_Skill1']=df['Generic_matchings'].str[1]
df['Generic_Skill2']=df['Generic_matchings'].str[3]
df['Generic_Skill3']=df['Generic_matchings'].str[5]
df['Generic_Skill4']=df['Generic_matchings'].str[7]
df['Generic_Skill5']=df['Generic_matchings'].str[9]
df['Generic_Skill6']=df['Generic_matchings'].str[11]
df['Generic_Skill7']=df['Generic_matchings'].str[13]
df['Generic_Skill8']=df['Generic_matchings'].str[15]
df['Generic_Skill9']=df['Generic_matchings'].str[17]
df['Generic_Skill10']=df['Generic_matchings'].str[19]
```

```python
#Preprocessing Job Types

df['Job Type']=df['Job Type'].replace('permanent','full-time')
df['Job Type']=df['Job Type'].fillna("full-time")
df['Job_Type']=df['Job Type'].str.split(",").str[0]
df['Job_Type']=df['Job_Type'].replace('full time','full-time')
```

```python
#Preprocessing Designation on the basis on Minimum Experience

df['Post']=df.Min_Experience.astype(int).apply(lambda x:'Entry Level' if x <= 2 else ('Mid-Senior Level' if (x>2 and x<=5) else 'Senior Level'))
```

```python
#Columns can be dropped as per requirement to extract dataframe for model building


#df=df.drop('Description',axis=1)
#df=df.drop('Job Type',axis=1)
#df=df.drop('Location',axis=1)
#df=df.drop('Company',axis=1)
#df=df.drop('Salary',axis=1)
#df=df.drop('Experience',axis=1)
#df=df.drop('SkillCount',axis=1)
#df=df.drop('Keywords',axis=1)
```

```python
#Label encoder example 
#Required features can be encoded as per the model

le=LabelEncoder()
df['Encoded_Location']= le.fit_transform(df['Location']) 
```
